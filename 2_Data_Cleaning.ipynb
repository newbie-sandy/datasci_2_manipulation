# Data Cleaning with Pandas:

## Identify and handle missing values in the dataset.

```python
# filling missing data represented using str"missing" with pd.NA
missing_data = data == 'missing'
df = data.copy()
df[missing_data] = pd.NA
```

```python
df
```

## Remove any duplicate rows and columns, if they exist.

Result: duplicate rows and column don't exist.

```python
# Identify duplicate columns
duplicate_columns = data.columns[data.columns.duplicated(keep='first')]
```

```python
# Print the duplicate columns
print(duplicate_columns)
```

```python
# Remove duplicate columns
data = data.loc[:, ~data.columns.duplicated(keep='first')]
```

```python
# Identify duplicate rows
duplicate_rows = data[data.duplicated(keep='first')]
```

```python
# Print or inspect the duplicate rows
print(duplicate_rows)
```

```python
# Remove duplicate rows
data = data.drop_duplicates(keep='first')
```

# Data Transformation:


## Create new columns based on existing ones (e.g., if you have a 'birth_date' column, create an 'age' column).

Created a new "Weight (in lb)" column based on "Weight (in kg)"

```python
df.dtypes
```

```python
# Convert the column to numeric, coercing invalid values to NaN
df['Weight (in kg)'] = pd.to_numeric(df['Weight (in kg)'],errors= 'coerce',downcast='integer')
```

```python
# Create column Weight (in lb)
df['Weight (in lb)'] = df['Weight (in kg)'] * 2.20462
```

```python
# Insert the new column after the exisiting one
column_position = df.columns.get_loc('Weight (in kg)') + 1  # Get the index of 'W in kg' and add 1 to place it after
df.insert(column_position, 'Weight (in lb)', df.pop('Weight (in lb)'))  # Pop the column and insert it in the desired position
```

```python
import modin.pandas as mpd
import pandas as pd
import time

# Define the dataset path
dataset_path = 'synthea_1m_fhir_3_0_May_24.tar.gz'

# Using Modin
start_time = time.time()
modin_df = mpd.read_csv(dataset_path)
modin_filtered = modin_df[modin_df['condition'] == 'Heart Disease']
modin_average_age = modin_filtered['age'].mean()
modin_time = time.time() - start_time

# Using Pandas
start_time = time.time()
pandas_df = pd.read_csv(dataset_path)
pandas_filtered = pandas_df[pandas_df['condition'] == 'Heart Disease']
pandas_average_age = pandas_filtered['age'].mean()
pandas_time = time.time() - start_time

# Print results and execution times
print(f"Modin Average Age of Patients with Heart Disease: {modin_average_age:.2f} years")
print(f"Pandas Average Age of Patients with Heart Disease: {pandas_average_age:.2f} years")
print(f"Modin Execution Time: {modin_time:.4f} seconds")
print(f"Pandas Execution Time: {pandas_time:.4f} seconds")
```
